# Checking how different variables are related to our profit

```python
# Profit vs State
plt.figure(figsize=(35,15))
plt.bar(df['State'],df['Profit'])
plt.xlabel('State')
plt.ylabel('Profit')
plt.title('Profit Vs State')
plt.xticks(rotation=90)
plt.show()
```

The graph shows the range of profit made my each state. From here we can see that North Carolina ,Texas,Illinoise Chino are the states making severe losses.

Let's find out the total profit a state is making

```python
#Total Profit vs States
plt.figure(figsize=(20,10))
state=df.groupby(["State"])["Profit"].sum()
state.sort_values(ascending=False,inplace=True)
state.plot.bar()
plt.show()
```

California has the highest profit. The states like New York , Washington etc are showing good profit.But some states like Texas , Ohio are incurring overall loss even though the paticipation was more in these states.

```python
#Total Profit vs Ship Mode
plt.figure(figsize=(10,5))
ship=df.groupby(["Ship Mode"])["Profit"].sum()
ship.sort_values(ascending=False,inplace=True)
ship.plot.bar()
plt.show()
```

Standard class is showing highest total profit and it was also preffered by most of our customers. Same Day Delivery is making least total profit, this maybe because they also has least count among the ship modes. All the ship modes has positive total profit. This means the company is definitely making overall profit. And that overall profit is the sum of all the total profits made by each ship mode.

```python
#Total Profit vs Segment
plt.figure(figsize=(10,5))
segment=df.groupby(["Segment"])["Profit"].sum()
segment.sort_values(ascending=False,inplace=True)
segment.plot.bar()
plt.show()
```

Consumer Segment is making the highest total profit followed by Corporate. 

```python
#Total Profit vs City  
plt.figure(figsize=(10,5))
data=df.groupby(["City"])["Profit"].sum()
df1=data[data>10000]
df1.sort_values(ascending=False,inplace=True)
df1.plot.bar()#plotting the graphs of only those cities with total profit more than 10000
plt.show()
plt.figure(figsize=(10,5))
df2=data[data<-2000]
df2.sort_values(ascending=False,inplace=True)
df2.plot.bar()#plotting the graphs of only those cities with total profit less than -2000
plt.show()
```

The most profit making cities are New York City , Los Angeles and Seattle. Philandelphia,Huston,San Antonio,Lancster and Chicago are makin huge losses.

```python
#Total Profit Vs Category 
plt.figure(figsize=(10,5))
category=df.groupby(["Category"])["Profit"].sum()
category.sort_values(ascending=False,inplace=True)
category.plot.bar()
plt.show()
```

Technology is making maximum total profit despite of having lowest counts. This is because per unit technology selling is the most profitable thing. 

```python
#Total Profit Vs Sub-Category
plt.figure(figsize=(10,5))
scategory=df.groupby(["Sub-Category"])["Profit"].sum()
scategory.sort_values(ascending=False,inplace=True)
scategory.plot.bar()
plt.show()
```

Copiers are making maximum total profit despite of having lowest count as it belongs to technology category. It means our company should work on the sales of our copiers . 

```python
#Profit and sales 
plt.figure(figsize=(10,5))
sns.scatterplot(df['Sales'],df['Profit'])
```

We can see that upto a certain point with the increase in sales  profit increases but after that there comes a range of sales where increasing sales is not helping much with profit. But once that range is crossed increasing sales, increases profit. So, we must try that our total sales shouldn't fall in that uncertain range.

```python
#Total Profit vs Region 
plt.figure(figsize=(10,5))
region=df.groupby(["Region"])["Profit"].sum()
region.sort_values(ascending=False,inplace=True)
region.plot.bar()
plt.show()
```

West is making maximum total profit and also had maximum count. While central region is struggling in both total profit and counts. We need to focus on central and south region.

```python
#heatmap to show the correlation of the dataset
plt.figure(figsize=(10,5))
sns.heatmap(df.corr(),annot=True)
plt.show()
```

The discount is inversely proportional to the profit as well as sales, so the idea of increasing discounts to increase sales or sales quantity can be discarded easily. The correlation between Profit and sales is significant .With the increase in sales, profit increases. There is also positive correlation between sales and quantity.
